```python
# Handling the Outliers
```

```python
#find the outliers and replace them by median
for col_name in raw_data1.columns[:-1]:
    q1 = raw_data1[col_name].quantile(0.25)
    q3 = raw_data1[col_name].quantile(0.75)
    iqr = q3 - q1
    
    low = q1-1.5*iqr
    high = q3+1.5*iqr
    
    raw_data1.loc[(raw_data1[col_name] < low) | (raw_data1[col_name] > high), col_name] = raw_data1[col_name].median()
```

```python


def click3():
        print("Regression Model")
        print("Model Training Started")
        global model
        X=raw_data.iloc[:,0:11]
        y = raw_data.iloc[:,11]
        # Initialize Sequential model
        
        from scipy.stats import zscore

        # Scaling the independent attributes using zscore
        X=X.apply(zscore)
        
        model = tensorflow.keras.models.Sequential()

        # Normalize input data
        model.add(tensorflow.keras.layers.BatchNormalization(input_shape=(11,)))

        # Add final Dense layer for prediction - Tensorflow.keras declares weights and bias automatically
        model.add(tensorflow.keras.layers.Dense(1))

        model.compile(optimizer='sgd', loss='mean_squared_error')

        model.fit(X, y, epochs=200, validation_split=0.35)

        Label(window,text="Network Trained",width=20).grid(row=3,column=2,sticky=E)

        print("Trained")
def click4():
    model.save("my_model")
    Label(window,text="Saved model to disk",width=20).grid(row=4,column=2,sticky=E)
    print("Model Saved")
```

```python
def click5():
    
    print("Classification Model")
    print("Model Training Started")
    global model1
    #independent and dependent variables
    X=raw_data.iloc[:,0:11]
    y = raw_data.iloc[:,11]
    from scipy.stats import zscore

    # Scaling the independent attributes using zscore
    X_z=X.apply(zscore)
    
    
    from sklearn.model_selection import train_test_split

    X_train, X_test, y_train, y_test = train_test_split(X_z,y)
    
    from tensorflow.keras.utils import to_categorical

    y_train = to_categorical(y_train, num_classes=9)
    y_test = to_categorical(y_test, num_classes=9)

    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense
    from tensorflow.keras import regularizers, optimizers

    ## hyperparameters
    iterations = 100
    learning_rate = 2e-1
    hidden_nodes = 256
    output_nodes = 9
    Lambda = 1e-4

    model1 = Sequential()
    model1.add(Dense(hidden_nodes, input_shape=(11,), activation='relu'))    
    model1.add(Dense(hidden_nodes, activation='relu'))
    model1.add(Dense(output_nodes, activation='softmax', kernel_regularizer=regularizers.l2(Lambda)))

    sgd = optimizers.SGD(lr=learning_rate, decay=1e-6, momentum=0.99)
    # Compile model
    model1.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

    # Fit the model
    model1.fit(X_train, y_train, epochs=iterations, batch_size=1000, verbose= 1)
    
    Label(window,text="Network Trained",width=20).grid(row=6,column=2,sticky=E)
    print("Trained")
    
    
def click6():
    model1.save("my_model1")
    Label(window,text="Saved Model to disk",width=20).grid(row=7,column=2,sticky=E)
    print("Model Saved")
```
